##**Column Renaming => Task -1**

```python
# Task 1: Rename the Column names
def rename_columns(df):
    column_mapping = {
        'State name': 'State/UT',
        'District name': 'District',
        'Male_Literate': 'Literate_Male',
        'Female_Literate': 'Literate_Female',
        'Rural_Households': 'Households_Rural',
        'Urban_ Households': 'Households_Urban',
        'Age_Group_0_29': 'Young_and_Adult',
        'Age_Group_30_49': 'Middle_Aged',
        'Age_Group_50': 'Senior_Citizen',
        'Age not stated': 'Age_Not_Stated'
    }
    print(f"Original columns: {df.columns.tolist()}")
    df.rename(columns=column_mapping, inplace=True)
    print(f"Renamed columns: {df.columns.tolist()}")
    return df
```

###**Task-2 : Rename State/UT Names**

```python
# Task 2: Rename State/UT Names
def rename_state_ut_names(df):
    if 'State/UT' in df.columns:
        df['State/UT'] = df['State/UT'].apply(lambda x: ' '.join([word.capitalize() if word != 'and' else word.lower() for word in x.split()]))
        print("State/UT names standardized successfully")
    else:
        print("Error: 'State/UT' column not found in DataFrame")
    return df
```

**Task-3: New State/UT formation with encoding handeling**

- In 2014 Telangana was formed after it split from Andhra Pradesh, The districts that were included in Telangana are stored in Data/Telangana.txt . Read the text file and Rename the State/UT From “Andhra Pradesh” to “Telangana” for the given districts.

- In 2019 Ladakh was formed after it split from Jammu and Kashmir, which included the districts Leh and Kargil.  Rename the State/UT From “Jammu and Kashmir” to “Ladakh” for the given districts.

```python
# Task 3: New State/UT formation with encoding handling
def update_new_state_ut(df, telangana_file, ladakh_districts):
    try:
        with open(telangana_file, 'r', encoding='utf-8') as file:
            telangana_districts = file.read().splitlines()
    except UnicodeDecodeError:
        with open(telangana_file, 'r', encoding='latin-1') as file:
            telangana_districts = file.read().splitlines()

    print(f"Telangana districts: {telangana_districts}")
    print(f"Ladakh districts: {ladakh_districts}")

    if 'District' in df.columns:
        print(f"Sample 'District' values before update: {df['District'].unique().tolist()[:10]}")  # Show sample data for verification
        df.loc[df['District'].isin(telangana_districts), 'State/UT'] = 'Telangana'
        df.loc[df['District'].isin(ladakh_districts), 'State/UT'] = 'Ladakh'
        print(f"Sample 'District' values after update: {df['District'].unique().tolist()[:10]}")  # Show sample data for verification
        print("New State/UT formations updated successfully")
    else:
        print("Error: 'District' column not found in DataFrame")
        print(f"Current columns: {df.columns.tolist()}")
    return df
```

```python
df = update_new_state_ut(df, '/content/drive/MyDrive/Soumya/Telangana.docx', ['Leh(Ladakh)', 'Kargil'])
```

###**Task-4 : Find and process Missing Data**

```python
# Task 4: Find and process Missing Data
def handle_missing_data(df):
    # Calculate and store the percentage of missing data for each column before filling
    missing_data_before = df.isnull().mean() * 100
    print("Missing data before filling:")
    print(missing_data_before)

    # Filling missing data based on provided hints
    if 'Population' in df.columns and 'Male' in df.columns and 'Female' in df.columns:
        df['Population'] = df['Population'].fillna(df['Male'] + df['Female'])

    if 'Literate' in df.columns and 'Literate_Male' in df.columns and 'Literate_Female' in df.columns:
        df['Literate'] = df['Literate'].fillna(df['Literate_Male'] + df['Literate_Female'])

    if 'Population' in df.columns and 'Young_and_Adult' in df.columns and 'Middle_Aged' in df.columns and 'Senior_Citizen' in df.columns and 'Age_Not_Stated' in df.columns:
        df['Population'] = df['Population'].fillna(df['Young_and_Adult'] + df['Middle_Aged'] + df['Senior_Citizen'] + df['Age_Not_Stated'])

    if 'Households' in df.columns and 'Households_Rural' in df.columns and 'Households_Urban' in df.columns:
        df['Households'] = df['Households'].fillna(df['Households_Rural'] + df['Urban_Households'])

    # Calculate and store the percentage of missing data for each column after filling
    missing_data_after = df.isnull().mean() * 100
    print("Missing data after filling:")
    print(missing_data_after)

    return df, missing_data_before, missing_data_after
```

```python
df, missing_data_before, missing_data_after = handle_missing_data(df)
```

###**Analayz & Visualize the Data**

```python
# Analysis Task 1: Total population of each district
total_population_district = df.groupby('District')['Population'].sum().reset_index()
total_population_district.sort_values(by='Population', ascending=False, inplace=True)

plt.figure(figsize=(12, 8))
sns.barplot(x='Population', y='District', data=total_population_district.head(10))
plt.title('Top 10 Districts by Total Population')
plt.xlabel('Total Population')
plt.ylabel('District')
plt.show()
```

```python
# Analysis Task 2: Literate males and females in each district
literate_district = df.groupby('District')[['Literate_Male', 'Literate_Female']].sum().reset_index()

plt.figure(figsize=(12, 8))
literate_district = literate_district.melt(id_vars=['District'], value_vars=['Literate_Male', 'Literate_Female'], var_name='Gender', value_name='Count')
sns.barplot(x='Count', y='District', hue='Gender', data=literate_district.head(20))
plt.title('Literate Males and Females in Top 20 Districts')
plt.xlabel('Count')
plt.ylabel('District')
plt.show()
```

```python
# Analysis Task 3: Percentage of workers in each district
df['Worker_Percentage'] = (df['Male_Workers'] + df['Female_Workers']) / df['Population'] * 100
worker_percentage_district = df.groupby('District')['Worker_Percentage'].mean().reset_index()

plt.figure(figsize=(12, 8))
sns.barplot(x='Worker_Percentage', y='District', data=worker_percentage_district.sort_values(by='Worker_Percentage', ascending=False).head(10))
plt.title('Top 10 Districts by Worker Percentage')
plt.xlabel('Worker Percentage')
plt.ylabel('District')
plt.show()
```

```python
# Visualization Task: Total population by state
total_population_state = df.groupby('State/UT')['Population'].sum().reset_index()
total_population_state.sort_values(by='Population', ascending=False, inplace=True)

plt.figure(figsize=(14, 10))
sns.barplot(x='Population', y='State/UT', data=total_population_state)
plt.title('Total Population by State/UT')
plt.xlabel('Total Population')
plt.ylabel('State/UT')
plt.show()
```

```python
# Visualization Task: Literacy rate by state
df['Literacy_Rate'] = (df['Literate_Male'] + df['Literate_Female']) / df['Population'] * 100
literacy_rate_state = df.groupby('State/UT')['Literacy_Rate'].mean().reset_index()

plt.figure(figsize=(14, 10))
sns.barplot(x='Literacy_Rate', y='State/UT', data=literacy_rate_state.sort_values(by='Literacy_Rate', ascending=False))
plt.title('Literacy Rate by State/UT')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('State/UT')
plt.show()
```
